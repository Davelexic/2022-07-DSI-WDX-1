```sql
/*
Desplay invoice data. 
*/
SELECT 
    *
FROM
    Invoices
```

```sql
/*
Show invoices that sitll have a positive credit total, show the vendor ID anad Name.
*/
SELECT
     invoices.InvoiceNumber 
    ,invoices.CreditTotal
    ,Vendors.VendorID
    ,Vendors.VendorContactFName
    ,Vendors.VendorContactLName
FROM
    Invoices
JOIN
    Vendors
ON
    Vendors.VendorID=Invoices.VendorID
WHERE
    CreditTotal
    > 0
```

```sql
/*
Show the Full Names of vendors and their Invocice Average that is greater that $1,500.
*/
SELECT
     Vendors.VendorID
    ,Vendors.VendorContactFName
    ,Vendors.VendorContactLName
    ,AVG(invoices.InvoiceTotal) AS AverageInvoiceTotal
FROM
    Invoices
JOIN
    Vendors
ON
    Vendors.VendorID=Invoices.VendorID
GROUP BY
     Vendors.VendorID
    ,Vendors.VendorContactFName
    ,Vendors.VendorContactLName
HAVING
    AVG(invoices.InvoiceTotal)
    >1500
ORDER BY
    vendors.VendorContactLName
```

```sql
/*
Show all Companies with Invoices in 2020.
*/
SELECT
     vendors.VendorName
    ,invoices.InvoiceDate
FROM
    Vendors
JOIN
    Invoices
ON
    vendors.VendorID=invoices.VendorID
WHERE
     invoices.InvoiceDate between '2019-12-31' and '2021-01-01'
ORDER BY
    vendors.VendorName
```

```sql
/*
Show all Companies who have not made a payment and their invice date, sort by date.
*/
SELECT
      vendors.VendorName
     ,invoices.InvoiceDate
     ,PaymentDate
FROM
    Vendors
JOIN
    Invoices
ON
    vendors.VendorID=invoices.VendorID
WHERE
    invoices.PaymentDate
IS
    NULL
ORDER BY
    InvoiceDate
```
